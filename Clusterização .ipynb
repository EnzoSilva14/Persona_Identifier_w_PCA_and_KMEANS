```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=data.copy()
num_rows = df.shape[0]
bins = int(np.sqrt(num_rows))

def plot_column(df, column):
    plt.figure()
    df[column].plot.hist(bins=bins)
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Frequencia')
    plt.show()
```

```python
# Removendo as colunas que não fazem sentido para a análise
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 5', 'Unnamed: 11', 'Unnamed: 16', 'Unnamed: 19', 'Unnamed: 23',
                      'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 38', 'Unnamed: 44', 'Unnamed: 48', 'Unnamed: 52',
                      'Unnamed: 59', 'Unnamed: 64', 'Unnamed: 71'])
```

```python
# Verificando valores faltantes
print(df.isnull().sum())

# Removendo linhas com valores faltantes
df.dropna(inplace=True)
```

```python
# Loop para remover espaços em branco das colunas
for col in df.columns:
    df.rename(columns={col: col.replace(' ', '')}, inplace=True)
    df.rename(columns={col: col.replace('.', '')}, inplace=True)
```

```python
# Selecionando colunas que começam com "P" utilizando list comprehension
cols_to_convert = [col for col in df.columns if col.startswith('P')]

# Convertendo as colunas selecionadas para o tipo float
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
#linhas estranhas
for col in df.columns:
    if col.startswith('P'):
        plot_column(df,col)
```

```python
from sklearn.cluster import KMeans
# Selecionando as variáveis numéricas
X = df.copy().drop(columns=['Qtipcama', 'Qche', 'Qclasse', 'Qs','Qciv'])

X
```

# TRUQUE DO COTOVELO

```python
# Normalizando as variáveis
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

# Seleciona as colunas com valores numéricos
y = X.select_dtypes(include=['float64', 'int64'])

# Testa diferentes valores de k
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(y)
    inertias.append(kmeans.inertia_)


# Definindo o número de clusters
kmeans = KMeans(n_clusters=4)

# Aplicando o modelo de clusterização
kmeans.fit(X_norm)

# Adicionando os clusters ao DataFrame
df['cluster'] = kmeans.labels_

# Plota o gráfico da métrica de qualidade em função de k
plt.plot(range(1, 10), inertias, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('Inércia')
plt.grid()
plt.show()


```

# APLICANDO PCA

```python
# Normalizando as variáveis
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# seleciona as colunas numéricas para a redução de dimensionalidade
y = X.select_dtypes(include=['float64', 'int64'])

# cria uma instância do PCA com 2 componentes
pca = PCA(n_components=2)

# aplica o PCA aos dados
y_pca = pca.fit_transform(y)

# cria um novo DataFrame com as duas componentes principais
X_pca = pd.DataFrame(y_pca, columns=['componente1', 'componente2'])


plt.scatter(y_pca[:, 0], y_pca[:, 1])
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.show()
```
